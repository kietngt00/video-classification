```python
import torch
from torchvision.io import (read_video, write_video)
import os
from tqdm import tqdm

def extract_frames(i):
    folder = './train' + str(i)
    tbar = tqdm(os.listdir(folder))
    for file in tbar:
        video = read_video(os.path.join(folder, file), pts_unit='sec', output_format='TCHW')[0]
        T = video.shape[0]
        if T == 0: continue
        idx = torch.linspace(0, T-1, 32).long()
        write_video(os.path.join("./save", file), torch.einsum('tchw->thwc', video[idx]), fps=30)
    
```

```python
with open("k400_train_path.txt", mode="r") as f:
    urls = f.readlines()
```

```python
def process(url, i):
    file_name = url.split('/')[-1][:-1]
    folder = "train" + str(i)
    
    !wget -c $url
    !mkdir $folder
    !tar zxf $file_name -C $folder
    
    extract_frames(i)
    print('Finish ', file_name, '\n')
    
    !rm $file_name
    !rm -rf $folder
```

```python
from multiprocessing import Process

for i in range(132, 152, 4):
    procs = [Process(target=process, args=(urls[i+j], j)) for j in range(4)]
    for j in range(4):
        procs[j].start()
    for j in range(4):
        procs[j].join()
    
```
